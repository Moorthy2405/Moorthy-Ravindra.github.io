```python
df.balance.map(float)
conv= {
    "age": int,"job":str,"marital":str,"education":str,"default":str,"balance":float,"housing":str,"loan":str,"contact":str,
    "day":float,"month":str,"duration":float,"campaign":int,"pdays":int,"previous":int,"poutcome":str,"y":str
}
def data_type_converstion(df,conv):
    for col, data_type in conv.items():
            df[col]=df[col].astype(data_type)
    return df
df1=data_type_converstion(df.copy(),conv)
updated_df=df1.dtypes
print(updated_df)
```

### 3. Find the average balance of the customer who belongs to the subscribed customer and non-subscribed customer and also use a related plot to show them in visualization.

```python
subscribed = df1[df1['y'] == 'yes']['balance'].mean()

non_subscribed = df1[df1['y'] == 'no']['balance'].mean()

labels = ['Subscribed', 'Non-Subscribed']
avg_balances = [subscribed,subscribed]

plt.bar(labels, avg_balances, color=['green', 'orange'])
plt.xlabel('Customer Category')
plt.ylabel('Avg Balance')
plt.title('Average Customer Subscription')
plt.tight_layout()
plt.show()


print("Avg Balance Subscribed:",subscribed)
print("Avg Balance - Non-Subscribed:",non_subscribed)
```
